```python
from datetime import datetime

# CLEANUP: Clean up the data
# (1) First two columns are interpreted as a Python date. Third column could be cleaned up.
# (2) Cash value columns are turned into integers
# (3) Yes/No columns are turned into booleans
def clean_cell(index,value):
    if index in (0,1):
        return datetime.strptime(value,'%d/%m/%Y').date()
    if index in (3,4,5):
        ival = int(value.replace('.0',''))
        # Although a decimal is specified: It is never used.
        assert ival==float(value), value
        return ival
    if index in (7,8):
        if value.lower()=='yes': return True
        if value.lower()=='no' : return False
        raise ValueError(value)
    return value
data = [rawdata[0]] + [ [clean_cell(i,value) for i,value in enumerate(row)] for row in rawdata[1:] ]

# Show me the first row cleaned up
len(data),data[1]
```

```python
import string

# EXPLORING: Which columns are discrete values? Look for set size of column's values
def setSize(columnIndex):
    values = [ row[columnIndex] for row in data[1:]]
    return len(set(values))
[ (string.uppercase[i],header,setSize(i)) for i,header in enumerate(data[0]) ]
```

```python
from collections import Counter

# EXPLORING: Big clever summary of common values for each column
def summary(columnIndex):
    values = [ row[columnIndex] for row in data[1:] ]
    values = Counter(values)
    return values.most_common(5)
# For each column:
#  * Excel column letter (A-Z)
#  * Column title
#  * Top five popular values. Useless for continuous data like cash, dates etc.
[ (string.uppercase[i],header,setSize(i),summary(i),'--------') for i,header in enumerate(data[0]) ]
```

```python
# CLEANUP: Create a proper list of key->value dicts
datahash = [ { data[0][i]:value for i,value in enumerate(row) } for row in data[1:] ]
datahash[0]
```

```python
# CLEANUP: Calculate a clean Continent per row
def get_continent(row):
    location = row['Location Of Investee'].lower()
    geo = row['Geography Of Beneficiaries'].lower()
    if location=='africa': return 'Africa'
    if geo=='asia'  : return 'Asia'
    if location=='india' : return 'Asia'
    if location=='uk' or location[:3]=='uk,'  : return 'Europe'
    if location=='canada': return 'North America'
    if geo=='europe': return 'Europe'
    if location=='' and geo=='': return '(unknown)'
    raise ValueError(location,geo)
# Test it works
print set([ get_continent(row) for row in datahash ])
# Add to the hash
for row in datahash:
    row['continent'] = get_continent(row)
```

```python
# CLEANUP: As above. Add a country to each row.
def get_country(row):
    location = row['Location Of Investee'].lower()
    geo = row['Geography Of Beneficiaries'].lower()
    if location=='uk': return 'UK'
    if location[:3]=='uk,': return 'UK'
    if location=='india': return 'India'
    if location=='canada': return 'Canada'
    if geo=='africa': return '(unknown)'
    if geo=='asia': return '(unknown)'
    if geo=='europe': return '(unknown)'
    if geo=='': return '(unknown)'
    raise ValueError(location,geo)
# --
# These results appear to be useless
print set([ get_country(row) for row in datahash ])
for row in datahash:
    row['country'] = get_country(row)
```

```python
# CLEANUP: Calculate a year for each row
for row in datahash:
    row['year'] = row['Investment date (commitment for\ninvestment/ facility/ guarantee)'].year
```

```python
 # CLEANUP: Actual name of Foundation (Source: Google, background reading)
foundations = {
               'tn9 1ap' : 'Panahpur',
               'UK,TN91AP' : 'Panahpur',
               'WC1R 4BH ': 'The LankellyChase Foundation',
               'UK,WC1R4BH': 'The LankellyChase Foundation',
               'WC2B4AS': 'The Barrow Cadbury Trust',
               'UK,WC2B4AS': 'The Barrow Cadbury Trust'
               }
for row in datahash:
    row['foundation'] = foundations[row['Source Of Investment']]
```

```python
# CLEANUP: Add a cleaned 'Cash Invested' column for clarity
for row in datahash:
    row['Cash Invested'] = row['Cash invested - principal invested/ amount guaranteed\nor underwritten']
```

```python
import json

# EXPORT: Construct a SanKey diagram
# --
# Interesting fields:
#   Type of organisation
#   Product Type
#   Purpose of Investment
#   Sector
# --
# Output
# {"nodes":[  {"name":"Agricultural 'waste'"},... ],
#  "links":[  {"source":0,"target":1,"value":124.729,foundationSplit:[['Panahpur',9999]...]}, ..]}
nodes = []
links = {}
def create_links(column_from,column_to):
    for row in datahash:
        foundation = row['foundation']
        a = str(row[column_from])
        b = str(row[column_to])
        assert a,row
        assert b,row
        # Maintain the ordered list of nodes
        if a not in nodes:
            nodes.append(a)
        if b not in nodes:
            nodes.append(b)
        a_index = nodes.index(a)
        b_index = nodes.index(b)
        # Maintain the big hash of unique links
        link = links[(a_index,b_index)] = links.get((a_index,b_index),{})
        link['total'] = link.get('total',0) + row['Cash Invested']
        link[foundation] = link.get(foundation,0) + row['Cash Invested']
def foundationSplit(link):
    return [ x for x in link.items() if x[0]!='total']
#create_links('year','Product Type')
create_links('Type Of Organisation Invested In','Product Type')
create_links('Product Type','Purpose Of Investment')
create_links('Purpose Of Investment','Sector')

summary_sankey = { 
    'nodes': [{'name':x} for x in nodes], 
    'links': [ {
          'source':a_index,
          'target':b_index,
          'value':link['total'],
          'foundationSplit':foundationSplit(link),
          } 
        for ((a_index,b_index),link) in links.iteritems()
    ]
}
```

```python
# EXPORT: Construct a stacked bar chart
# --
# Render year-on-year investment by sector
def get_2013_total(fieldname,fieldvalue):
    out = 0
    for row in datahash:
        if row['year']==2013 and row[fieldname]==fieldvalue:
            out += row['Cash Invested']
    return out
            
major = 'year'
minor = 'Sector'
all_major_values = sorted(list(set([ x[major] for x in datahash if x[major]!=2008])))
all_minor_values = sorted(list(set([ x[minor] for x in datahash])), key=lambda x:get_2013_total(minor,x), reverse=True)
def get_series(major_value,foundation=None):
    totals = {}
    for row in datahash:
        if not row[major]==major_value: continue
        if foundation and not (row['foundation']==foundation): 
            continue
        minor_value = row[minor]
        totals[minor_value] = totals.get(minor_value,0) + row['Cash Invested']
    return { 
            'major': str(major_value), 
            'elements':[{'name':x,'value':totals.get(x,0)} for x in all_minor_values]
            }
# Multiple barcharts are constructed for the animated transitions
year_on_year_barchart = {}
year_on_year_barchart['all'] = { 'legend': all_minor_values, 'series':[ get_series(x) for x in all_major_values ] }
year_on_year_barchart['foundation_panahpur'] = { 'legend': all_minor_values, 'series':[ get_series(x,'Panahpur') for x in all_major_values ]}
year_on_year_barchart['foundation_lankelly'] = { 'legend': all_minor_values, 'series':[ get_series(x,'The LankellyChase Foundation') for x in all_major_values ]}
year_on_year_barchart['foundation_barrow'] = { 'legend': all_minor_values, 'series':[ get_series(x,'The Barrow Cadbury Trust') for x in all_major_values ]}
```

```python
# EXPORT: Construct a pie chart
# ---
# (1) Total per sector
sector_totals = {}
for row in datahash:
    sector = row['Sector']
    sector_totals[sector] = sector_totals.get(sector,0) + row['Cash Invested']
sector_names = sorted(sector_totals.items(),key=lambda x:x[1])
sector_names = [ x[0] for x in sector_names ]

def sector_totals(foundation=None):
    out = {}
    for row in datahash:
        if foundation and row['foundation']!=foundation:
            continue
        sector = row['Sector']
        value = row['Cash Invested']
        out[sector] = out.get(sector,0) + value
    return [{'name':sector_name,'value':out.get(sector_name,0)} for sector_name in sector_names]
# Like bar charts, multiple pie charts are created for animated transitions
sector_piechart = {
    'all' : sector_totals(),
    'foundation_panahpur' : sector_totals(foundation='Panahpur'),
    'foundation_lankelly' : sector_totals(foundation='The LankellyChase Foundation'),
    'foundation_barrow'   : sector_totals(foundation='The Barrow Cadbury Trust'),
}
sector_names
```

```python
# EXPORT: Construct a pie chart
# ---
# (2) Secured vs Unsecured
count = {}
for row in datahash:
    type = row['Product Type']
    count[type] = count.get(type,0) + row['Cash Invested']
def sortorder(x):
    # Look closely for the hack here
    if x[0]=='Loans and facilities - Unsecured': return -2
    if x[0]=='Loans and facilities - Partially secured': return -1
    return x[1]
product_types = sorted(count.items(),key=sortorder)
product_types = [ x[0] for x in product_types ]

def product_type_totals(foundation=None):
    out = {}
    for row in datahash:
        if foundation and row['foundation']!=foundation:
            continue
        type = row['Product Type']
        out[type] = out.get(type,0) + row['Cash Invested']
    return [{'name':type_name,'value':out.get(type_name,0)} for type_name in product_types]

unsecured_piechart = {
    'all' : product_type_totals(),
    'foundation_panahpur' : product_type_totals(foundation='Panahpur'),
    'foundation_lankelly' : product_type_totals(foundation='The LankellyChase Foundation'),
    'foundation_barrow'   : product_type_totals(foundation='The Barrow Cadbury Trust'),
}
```

```python
# EXPLORE: "Total Value" duplicates. Many rows refer to the same major coinvestment.
sum([x['Total Value incl. co-investment'] for x in datahash]), sum(set([x['Total Value incl. co-investment'] for x in datahash]))
```

```python
def sum_column(column_name,foundation=None):
    out = 0
    for x in datahash:
        if foundation and not (x['foundation']==foundation):
            continue
        out += x[column_name]
    return out
investment_total = { 
                      'all' : sum_column('Cash Invested'),
                      'foundation_panahpur': sum_column('Cash Invested','Panahpur'),
                      'foundation_lankelly': sum_column('Cash Invested','The LankellyChase Foundation'),
                      'foundation_barrow': sum_column('Cash Invested','The Barrow Cadbury Trust')
}
investment_total
```

```python
# CLEANUP: It appears that any duplicate "Total Value" entries are the same major investment, with lots of little investments inside.
# EXCEPT where Cash Invested == Total Value. (Those are not part of a coinvestment strategy).
# Extract a list of all coinvestments
coinvestments = []
for row in datahash:
    if row['Total Value incl. co-investment']!=row['Cash Invested']:
        # Tricky case: Try to append this to an existing coinvestment
        existing = [ x for x in coinvestments if x['value']==row['Total Value incl. co-investment'] ]
        if len(existing):
            existing[0]['investments'].append(row)
            continue
    # Fall-through: Create a new "major investment" row
    coinvestments.append({
        'value': row['Total Value incl. co-investment'],
        'investments': [row]
    })
len(datahash), len(coinvestments)
```

```python
# Create a sunburst hierarchy of coinvestment
def coinvestment_entries():
    for entry in coinvestments:
        total = sum([x['Cash Invested'] for x in entry['investments']])
        if entry['value']==total:
            for investment in entry['investments']:
                yield { 'name':investment['foundation'],'size':investment['Cash Invested'] }
        else:
            children = []
            tmp = {}
            for investment in entry['investments']:
                foundation = investment['foundation']
                cash = investment['Cash Invested']
                tmp[foundation] = tmp.get(foundation,0) + cash
            for (foundation,cash) in tmp.items():
                children.append({ 'name':foundation,'size':cash })
            children.append({'name':'(others)','size':entry['value']-total})
            yield { 'size':entry['value'], 'children':children, 'name':'Total Investment' }
sunburst = {
   'name' : 'Coinvestment',
   'children': sorted(list(coinvestment_entries()),key=lambda x:x['size'])
   }
```

```python
# Sum of coinvestments
coinvestment_total = sum([x['value'] for x in coinvestments])
```

```python
# EXPORT: Write to file
graphs = {
          'bar' : year_on_year_barchart,
          'pie1' : sector_piechart,
          'pie2' : unsecured_piechart,
          'sankey' : summary_sankey,
          'sunburst' : sunburst,
          'coinvestment_total' : coinvestment_total,
          'investment_total' : investment_total,
          }
with open('graphs.json','w') as f:
    json.dump(graphs,f)
```
