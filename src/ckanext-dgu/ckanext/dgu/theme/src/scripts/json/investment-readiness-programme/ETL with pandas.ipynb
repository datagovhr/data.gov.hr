```python
# Requirements need to be installed globally.
import ckanclient
import pandas
import requests
import numpy
# -- Pandas options
pandas.set_option('max_columns',99)
ckan = ckanclient.CkanClient(base_location="http://data.gov.uk/api")
```

```python
import urllib

# Download the "Incubator data" resource from each of these packages
def download_ckan_csv(pkg_name,resource_name,encoding='iso-8859-1',cleanup=None):
    pkg = ckan.package_entity_get(pkg_name)
    resources = {x['description'].lower() : x for x in pkg['resources']}
    resource = resources[resource_name.lower()]
    url = resource['url']
    response = urllib.urlopen(url)
    raw = unicode(response.read(), encoding)
    if cleanup:
        raw = cleanup(raw)
    return raw
# -- utility function
def clean_bethnal_green_data(x):
    # This looks like a control character which has slipped into the text.
    # Adjusting the encoding will not get rid of it
    return x.replace(u'ï¾',u'')
incubator_raw = [
                 download_ckan_csv('social-incubator-fund-data-wayra-unltd','Incubator data'),
                 download_ckan_csv('sinsif','Incubator data',encoding='utf-8'),
                 download_ckan_csv('social-incubator-fund-data-hub-launchpad','Incubator data'),
                 download_ckan_csv('sifbgv','Incubator data',cleanup=clean_bethnal_green_data)
]
# Preview data
from pprint import pprint
pprint([x[1100:1200] for x in incubator_raw])
```

```python
from StringIO import StringIO

# Load incubator data into pandas
incubator_headers = [
       'incubator', 'cash', 'date', 'cohort_id', 'cohort_name', 'cohort_start',
       'cohort_end', 'geo', 'num_interested', '_ignore1', '_ignore2',  'num_applications', 'source_applications',
       'source_applications2',
       'question1', 'question2', 'age_lead_applicants', 'age_winning_founders',
       'percent_male', 'percent_female', 'num_employees', 'percent_teams_one_female',
       'percent_teams_half_female',
       'num_beneficiaries', 'num_beneficiaries_increase_6months', 'num_countries',
       'num_investor_opportunities', 'total_new_investment', 'num_teams_on_icrf',
       'num_teams_on_other', 'num_teams_public_service_contracts', 'num_investors_engaged',
       'withdrawls'
]
# --
tmp = []
for csv_text in incubator_raw:
    df = pandas.read_csv(StringIO(csv_text),parse_dates=[2,5,6])
    # Trim excess empty columns
    while len(df.columns) > len(incubator_headers):
        del df[ df.columns[-1] ]
    df.columns = incubator_headers
    # Clean up the blank rows
    df = df[ df['incubator'].notnull() ]
    tmp.append(df)
incubator = pandas.concat(tmp,ignore_index=True)
## Clean up the money
def parse_money(x):
    if pandas.isnull(x):
        return x
    out = unicode(x,'utf-8')
    out = out[1:].replace(',','')
    out = int(out)
    return out
incubator['cash'] = incubator['cash'].map(parse_money)
incubator
```

```python
# Download the "Venture data" resource from each of these packages
venture_raw = [
                 download_ckan_csv('social-incubator-fund-data-wayra-unltd','Venture data'),
                 download_ckan_csv('sinsif','Venture data',encoding='utf-8'),
                 download_ckan_csv('social-incubator-fund-data-hub-launchpad','Venture data'),
                 download_ckan_csv('sifbgv','Venture data',cleanup=clean_bethnal_green_data)
]
# Preview data
from pprint import pprint
pprint([x[1100:1200] for x in venture_raw])
```

```python
# Load venture data into pandas
venture_headers = ['venture', 'target', 'target2', 'sector', 
 'sector2', 'invested_type', 'invested']
# --
tmp = []
for csv_text in venture_raw:
    df = pandas.read_csv(StringIO(csv_text))
    # Trim excess empty columns
    while len(df.columns) > len(venture_headers):
        del df[ df.columns[-1] ]
    df.columns = venture_headers
    # Clean up the blank rows
    df = df[ df['venture'].notnull() ]
    tmp.append(df)
tmp[0]['incubator'] = 'wayra'
tmp[1]['incubator'] = 'north'
tmp[2]['incubator'] = 'hub'
tmp[3]['incubator'] = 'bethnal'
venture = pandas.concat(tmp,ignore_index=True)
venture['invested'] = venture['invested'].map(parse_money)
venture.tail()
```

```python
## LOAD ICRF DATA
##
package_name = 'icrf-data'
pkg = ckan.package_entity_get(package_name)
resources = filter(lambda x:x['resource_type']=='file',pkg['resources'])
url = resources[0]['url']
headers = ['name','cash','postcode']
icrf = pandas.read_excel(urllib.urlopen(url),0,names=headers,skip_footer=1)
#icrf = pandas.read_excel('data/ICRF Grants 31.01.14.xls',0,names=headers,skip_footer=1)
icrf.tail()
```

```python
## DATA CLEANUP: 
## Sparsely populated sectors can usually be merged...
sectors_merged = {
     'Asset Management & Managed Workspace':'Housing and local facilities',
#     'Family, friends and relationships'   :'Mental health and wellbeing',
#     'Income and financial inclusion'      :'Employment, training and education',
}
venture['sector_clean'] = venture.sector.map( lambda x:sectors_merged.get(x,x) )
venture['sector_clean'].value_counts()
```

```python
## Ordered list of "Target" values for the second pie chart
target_list = list(venture.target.value_counts().index)
## --
## Cleaned, ordered list of "Sector" values for the first pie chart
sector_list = list(venture.sector_clean.value_counts().index)
## --
## Icon per sector
icon_map = {
    u'Arts, heritage, sport and faith': 'icon-picture',
    u'Employment, training and education': 'icon-trophy',
    u'Environmental': 'icon-leaf',
    u'Family, friends and relationships': 'icon-group',
    u'Housing and local facilities': 'icon-home',
    u'Income and financial inclusion': 'icon-money',
    u'Mental health and wellbeing': 'icon-medkit',
    u'Physical health': 'icon-heart'
}
def get_icons(df):
    counts = df.sector_clean.value_counts()
    for key in sector_list:
        for n in range(counts.get(key,0)):
            icon = icon_map.get(key,'icon-question')
            yield icon
list(get_icons(venture[ venture['incubator']=='wayra' ]))
```

```python
# RENDER PIE CHART. Data structure for a chart:
# [{'name':'foo','value':123}, {'name'...} ...]
# -- This can be used to render a pie chart of sectors or targets
def pandas_pie(frame, column, reindex):
    tmp = frame.fillna(0).groupby([column])\
        .venture.count()\
        .reindex(index=reindex,fill_value=0)
    return [ {'name':x,'value':y} for (x,y) in tmp.iteritems() ]
# Demo on the WAYRA data
pandas_pie(venture[ venture['incubator']=='wayra' ], 'sector_clean', sector_list)
```

```python
## Data export
def icrf_dict(x):
    return {'name':x[1][0], 'amount':x[1][1], 'url':x[1][3]}
def sib_dict(name,df,img,url):
    summary_rows = incubator[incubator['incubator']==name]
    mean = df.invested.mean()
    withdrawls = summary_rows.withdrawls.sum()
    funding = summary_rows.cash.sum()
    to_int = lambda x: -1 if numpy.isnan(x) else int(x)
    return {
            'name':name,
            'total_funding':to_int(funding),
            'mean_investment':to_int(mean),
            'withdrawls':to_int(withdrawls),
            'investment_sectors':list(get_icons(df)),
            'img' : img, 
            'url' : url,
            'sector_pie':pandas_pie(df,'sector',sector_list),
            'target_pie':pandas_pie(df,'target',target_list),
            }
            
data = {
    'icrf_mean':icrf.cash.mean().round(),
    'icrf_count':int(icrf.cash.count()),
    'icrf_items':[ icrf_dict(x) for x in icrf.iterrows() ],
    'sib': [
        sib_dict( 'Bethnal Green Ventures', venture[venture['incubator']=='bethnal'], '/images/logo_bethnal.png', 'http://bethnalgreenventures.com/' ),
        sib_dict( 'Hub Launchpad', venture[venture['incubator']=='hub'], '/images/logo_hub.png', 'http://www.hublaunchpad.net/' ),
        sib_dict( 'Social Incubator North', venture[venture['incubator']=='north'], '/images/logo_socialincubatornorth.png', 'http://socialincubatornorth.org.uk/' ),
        sib_dict( 'Wayra UnLtd', venture[venture['incubator']=='wayra'], '/images/logo_wayra.png', 'http://wayra.org/unltd/' ),
    ],
    'icon_to_sector' : dict([(v,k) for k,v in icon_map.iteritems()]),
}
with open('investment-readiness-d3.json','w') as f:
    import json
    json.dump(data,f)
data
```
